### **INTRODUCTION**
Rice production is highly sensitive to climatic conditions, particularly temperature and precipitation. Understanding how long-term climate variability affects rice yield is essential for sustainable agricultural planning and food security. This study analyzes 30 years of climate and rice yield data to evaluate trends, relationships, and variability using data analytics tools.

---
#### **OBJECTIVES**
1. Examine long-term trends in rice yield, temperature, and precipitation
2. Determine the relationship between rice yield and climatic variables
3. Identify periods of high and low productivity
4. Quantify climate variability and its potential influence on yield
5. Visualize climate-yield interactions using Python and Power BI

---
##### **KEY FINDINGS**
- Rice yield is **more sensitive to precipitation than temperature**
- Climate variability explains significant yield fluctuations
- Stable temperature with unstable rainfall creates production risk
- Long-term yield trend shows gradual improvement with fluctuations

---
###### **OBSERVATIONS**
- High rainfall years do not always guarantee high yield, indicating other agronomic factors
- Low rainfall years frequently correspond with yield decline
- Climate adaptation strategies are necessary for yield stability

---
###### **CONCLUSION**
This analysis demonstrates that while temperature remains relatively stable over the study period, precipitation variability plays a crucial role in influencing rice yield. Effective water management, irrigation planning, and climate-smart agricultural practices are essential to mitigate yield risks under changing climatic conditions.

## ***Code behind my work***

```python
import pandas as pd
```

```python
crop_yield = pd.read_csv(r"c:\Users\DELL\Documents\bluetooth\FAOSTAT Rice data.csv")
crop_yield
```

```python
temperature = pd.read_csv(r"c:\Users\DELL\Documents\bluetooth\Mean annual temparature.csv")
temperature
```

```python
precipitation = pd.read_csv(r"c:\Users\DELL\Documents\bluetooth\Precipitation.csv")
precipitation
```

```python
# Merging crop yield data with temperature data
merge = pd.merge(crop_yield,temperature,left_on="Year",right_on="Category")
merge
```

```python
# Previous merging with Precipitation data
comb = pd.merge(merge,precipitation,on="Category")
comb
```

```python
# Fetching out the needed columns for the analysis
data = comb[["Year","Value","Average Mean Surface Air Temperature (°C)","Precipitation (mm)"]]
data
```

```python
# Renaming the needed columns
data.columns = ["Year","Rice_Yield","Temperature","Precipitation"]
data
```

```python
# Trend in Rice Yield Over Time
data[["Year","Rice_Yield"]].describe()
```

## **Insight:**
Rice yield shows inter-annual variability, with periods of increase and decline, indicating sensitivity to environmental factors

```python
# Long-term Climate Trend
data[["Temperature","Precipitation"]].mean()
```

## **Insight:**
- Mean annual temperature remains relatively stable
- Precipitation fluctuates more strongly, suggesting rainfall variability is a key risk factor

```python
# Correlation analysis
data[["Rice_Yield","Temperature","Precipitation"]].corr()
```

## **Insight:**
- __Rice yield shows__
    - Moderate correlation with precipitation
    - Weak correlation with temperature
>"This suggests rainfall availability has a stronger influence on yield than temperature alone

```python
# Highest and Lowest Yield Years
max = data.loc[data["Rice_Yield"].idxmax()]
print(max)
min = data.loc[data["Rice_Yield"].idxmin()]
print(min)
```

## **Insight:**
Extreme yield years often align with unusual precipitation patterns, reinforcing the importance of water availability

```python
# Climate Variability
data[["Temperature","Precipitation"]].std()
```

## **Insight:**
Precipitation variability is significantly higher than temperature variability, increasing production uncertainty.

## **Visualization: Rice Yield vs Climate**

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,8))
plt.plot(data["Year"], data["Rice_Yield"], label="Rice_Yield")
plt.plot(data["Year"], data["Temperature"], label="Temperature")
plt.plot(data["Year"], data["Precipitation"], label="Precipitation")
plt.legend()
plt.xlabel("Year")
plt.ylabel("Value")
plt.grid(True)
plt.title("Rice Yield, Temperature, and Precipitation (1993-2023)")
plt.show()
```
